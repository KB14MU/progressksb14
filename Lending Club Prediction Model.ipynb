# Setting up the Data

ready to set up our data for our Random Forest Classification Model!



```python
df.info()
loan = df
```

## Categorical Features




**Create a list of 1 element containing the string 'purpose'.

```python
cat_feats = pd.get_dummies(loan['purpose'],drop_first=True)
cat_feats.head()
```

```python
loan = loan.drop('purpose',axis =1)
loan.head()
```

```python
loan = pd.concat([loan,cat_feats],axis=1)
loan.head()
```

```python
X=loan.drop('not.fully.paid',axis=1)
y=loan['not.fully.paid']
```

```python
loan=loan.reset_index(drop=True)
loan.head()
```

## Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training a Decision Tree Model



```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit (X_train,y_train)
```

## Predictions and Evaluation of Decision Tree


```python
prediction = dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

## Training the Random Forest model



```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

## Predictions and Evaluation



```python
predict_rfc = rfc.predict(X_test)
```

**Now create a classification report from the results. Do you get anything strange or some sort of warning?**

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predict_rfc))
```

**Show the Confusion Matrix for the predictions.**

```python
print(confusion_matrix(y_test,predict_rfc))
```
